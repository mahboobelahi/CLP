```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
decriment=30
def plot_3d_rectangles(rectangles, x_ticks, y_ticks, z_ticks, step_size):
    plt.interactive(True)
    fig = fig = plt.figure(figsize=(20, 30))
    ax = fig.add_subplot(111, projection='3d')

    for rect in rectangles:
        x, y, z, length, width, height,_ = rect
        ax.bar3d(x, y, z, length, width, height,shade=True)


    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.set_title('3D Rectangles')

    ax.set_xticks(np.arange(0, x_ticks + step_size-decriment, step_size-decriment))
    ax.set_yticks(np.arange(0, y_ticks + step_size, step_size))
    ax.set_zticks(np.arange(0, z_ticks + step_size, step_size))
    plt.show()
```
