```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_rectangle(rectangles, x_ticks, y_ticks, z_ticks, step_size):
    plt.interactive(True)
    fig = plt.figure(figsize=(20, 30))
    ax = fig.add_subplot(111, projection='3d')
    
    for rect in rectangles:
        x, y, z, length, width, height, label = rect
        ax.bar3d(x, y, z, length, width, height, alpha=.4, edgecolor='black', lw=3)

        # Add label to the top face
        label_x = x + length / 2
        label_y = y + width / 2
        label_z = z + height

        ax.text(label_x, label_y, label_z, str(label), color='red', fontsize=12, ha='center', va='center')

        # Add label to the front face
        label_x_front = x + length / 2
        label_y_front = y + width
        label_z_front = z + height / 2

        ax.text(label_x_front, label_y_front, label_z_front, str(label), color='red', fontsize=12, ha='center', va='center')

    # Set ticks and labels
    ax.set_xticks(x_ticks)
    ax.set_yticks(y_ticks)
    ax.set_zticks(z_ticks)

    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.set_title('3D Rectangular Bars with Labels')

    plt.show()

# Example usage:
rectangles = [
    [0, 0, 0, 2, 3, 4, 10],
    [3, 3, 3, 1, 2, 3, 20],
    # Add more rectangles as needed
]

x_ticks = [0, 5, 10, 15, 20]
y_ticks = [0, 5, 10, 15, 20]
z_ticks = [0, 5, 10, 15, 20]
step_size = 1

plot_3d_rectangle(rectangles, x_ticks, y_ticks, z_ticks, step_size)
```

```python
def combine_spaces(space1, space2):
    # Extract coordinates for each space
    x1, y1, z1, l1, w1, h1 = space1
    x2, y2, z2, l2, w2, h2 = space2

    # Determine the new starting location and dimensions for the combined space
    x_combined = max(x1, x2)
    y_combined = y1  # Align with the y-coordinate of the last placed box
    z_combined = min(z1, z2)

    l_combined = max(x1 + l1, x2 + l2) - x_combined
    w_combined = w1  # Use the width of the last placed box
    h_combined = max(z1 + h1, z2 + h2) - z_combined

    # Create and return the combined space
    combined_space = [x_combined, y_combined, z_combined, l_combined, w_combined, h_combined]
    return combined_space

# Example spaces
beside_space = [0, 184, 0, 55, 51, 239]
front_space = [55, 92, 0, 26, 143, 239]

# Combine the "beside" and "front" spaces
combined_space = combine_spaces(beside_space, front_space)

# Print the result
print("Combined Space:", combined_space)
```
