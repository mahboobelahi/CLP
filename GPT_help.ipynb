### Ploting using matplotlib

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_rectangle(rectangles, x_ticks, y_ticks, z_ticks, step_size):
    plt.interactive(True)
    fig = plt.figure(figsize=(20, 30))
    ax = fig.add_subplot(111, projection='3d')
    
    for rect in rectangles:
        x, y, z, length, width, height, label = rect
        ax.bar3d(x, y, z, length, width, height, alpha=.4, edgecolor='black', lw=3)

        # Add label to the top face
        label_x = x + length / 2
        label_y = y + width / 2
        label_z = z + height

        ax.text(label_x, label_y, label_z, str(label), color='red', fontsize=12, ha='center', va='center')

        # Add label to the front face
        label_x_front = x + length / 2
        label_y_front = y + width
        label_z_front = z + height / 2

        ax.text(label_x_front, label_y_front, label_z_front, str(label), color='red', fontsize=12, ha='center', va='center')

    # Set ticks and labels
    ax.set_xticks(x_ticks)
    ax.set_yticks(y_ticks)
    ax.set_zticks(z_ticks)

    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.set_title('3D Rectangular Bars with Labels')

    plt.show()

# Example usage:
rectangles = [
    [0, 0, 0, 2, 3, 4, 10],
    [3, 3, 3, 1, 2, 3, 20],
    # Add more rectangles as needed
]

x_ticks = [0, 5, 10, 15, 20]
y_ticks = [0, 5, 10, 15, 20]
z_ticks = [0, 5, 10, 15, 20]
step_size = 1

plot_3d_rectangle(rectangles, x_ticks, y_ticks, z_ticks, step_size)
```
